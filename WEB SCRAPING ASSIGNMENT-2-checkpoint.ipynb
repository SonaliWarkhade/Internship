Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name,
experience_required. You have to scrape first 10 jobs data.

This task will be done in following steps:
    
1. first get the webpage https://www.naukri.com/
2. Enter “Data Analyst” in “Skill,Designations,Companies” field and enter “Bangalore” in “enter the location” field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.

Note- All of the above steps have to be done in code. No step is to be done manually.

```python
#after uploading it in jupyter notebook
driver=webdriver.Chrome("chromedriver.exe")
```

```python
#opening the homepage 
driver.get('https://www.naukri.com/')
#asking the user to input the keywords he/she wants to search
des = input('Enter the job you want to search :')
loc = input('Enter location you want to search :')
```

```python
search_des=driver.find_element_by_id("qsb-keyword-sugg")   #locating serach bar designation
search_loc=driver.find_element_by_id("qsb-location-sugg")  #locating search by location
search_des.clear()                                         #clearing search bar
search_loc.clear()
search_des.send_keys(des)                                  #sending user input to search bar
search_loc.send_keys(loc)

time.sleep(4)
search_button = driver.find_element_by_xpath('//div[@class="search-btn"]/button') #locating search bar
#clicking the button to start search
search_button.click()
```

```python
j_t=[] #job title
j_l=[] #job location
com_name=[] #company name
exp=[] #experience

for j in driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]'):
    j_t.append(j.text)
for m in driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi location"]/span'):
    j_l.append(m.text)
for k in driver.find_elements_by_xpath('//a[@class="subTitle ellipsis fleft"]'):
    com_name.append(k.text)
for l in driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi experience"]/span'):
    exp.append(l.text)
```

Let's create dataframe and store extracted data in table format:

```python
Naukari_Data_Analyst=pd.DataFrame({})
Naukari_Data_Analyst['Job_title']=j_t[0:10]
Naukari_Data_Analyst['Job_location']=j_l[0:10]
Naukari_Data_Analyst['Company_name']= com_name[0:10]
Naukari_Data_Analyst['Experience']=exp[0:10]
Naukari_Data_Analyst
```

```python
#fetching Urls
urls=[]
for j in driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]'):
    urls.append(j.get_attribute('href'))
    time.sleep(1) 
urls
```

```python
j_t=[] #job title
j_l=[] #job location
com_name=[] #company name
fj_d=[] #full job description
```

```python
a="no element found"
for i in urls:
    driver.get(i)
    
    try:
        for j in driver.find_elements_by_xpath('//h1[@class="jd-header-title"]'):
            j_t.append(j.text)
    except NoSuchElementException:
        j_t.append(a)
        
    try:
        for m in driver.find_elements_by_xpath('//span[@class="location "]'):
            j_l.append(m.text)
    except NoSuchElementException:
        j_l.append(a)
        
    try:
        for k in driver.find_elements_by_xpath('//section[@class="jd-header"]/div/div/div/a[1]'):
            com_name.append(k.text)
    except NoSuchElementException:
        com_name.append(a)
        
    try:
        for l in driver.find_elements_by_xpath('//section[@class="job-desc"]'):
            fj_d.append(l.text.replace('\n',' '))
            time.sleep(3)
    except NoSuchElementException:
        fj_d.append(a)
        
```

```python
#after uploading it in jupyter notebook
driver=webdriver.Chrome("chromedriver.exe")
```

```python
#opening the homepage of amazon.in
driver.get('https://www.amazon.in/')
#asking the user to input the keywords he/she wants to search
des = input('Enter the job you want to search :')
```

```python
search_des=driver.find_element_by_id("twotabsearchtextbox")   #locating serach bar 
search_des.clear()                                         #clearing search bar
search_des.send_keys(des)                                  #sending user input to search bar

time.sleep(4)
search_button = driver.find_element_by_xpath('//span[@class="nav-search-submit-text nav-sprite nav-progressive-attribute"]/input') #locating search bar
#clicking the button to start search
search_button.click()
```

```python
title=[]
rating=[]
price=[]
```

```python
try:
    for b in driver.find_elements_by_xpath('//span[@class="a-size-medium a-color-base a-text-normal"]'):
        title.append(b.text)
except NoSuchElementException:
    title.append(a)
    
try:
    for sd in driver.find_elements_by_xpath('//div[@class="a-row a-size-small"]/span[2]'):
        rating.append(sd.text)
except NoSuchElementException:
    rating.append(a)

try:
    for p in driver.find_elements_by_xpath('//span[@class="a-price"]/span/span[2]'):
        price.append(p.text)
except NoSuchElementException:
    price.append(a)
```
