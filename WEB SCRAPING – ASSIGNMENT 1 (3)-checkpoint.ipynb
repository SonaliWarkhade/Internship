3. Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e.
Name, IMDB rating, Year of release) and save it in form of a CSV file. 

```python
#importing the required libraries
from bs4 import BeautifulSoup
import requests
```

First we will send get request to the webserver to get the source code of the page which is to be scraped by using the request library

```python
page=requests.get("https://www.imdb.com/india/top-rated-indian-movies")
page
```

let's see the page content with the help of beautiful soup.

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify()) #see the content with alingment
```

```python
soup.title.text
```

```python
moives=soup.find_all('td', class_="titleColumn")
year=soup.find_all('span', class_='secondaryInfo')
rating=soup.find_all('td', class_="ratingColumn imdbRating")
```

```python
#extracting only 100 moive names
m=[]
for i in moives:
    title = i.a.text
    m.append(title)
    
moive_name= m[0:100]
moive_name
```

```python
#extracting year of release for 100 moives
y=[]
for i in year:
        y.append(i.text.replace('\n','')[1:-1] )

year_of_release= y[0:100]
year_of_release
```

```python
#extracting rating for 100 moives
r=[]
for i in rating:
        r.append(i.text.replace('\n',''))

IMDb_rating= r[0:100]
IMDb_rating
```

Let's create dataframe and store extracted data in table format:

```python
import pandas as pd
IMDBs_Top_rated_indian_movies=pd.DataFrame({})
IMDBs_Top_rated_indian_movies['moive_name']= moive_name
IMDBs_Top_rated_indian_movies['year_of_release']=year_of_release
IMDBs_Top_rated_indian_movies['IMDb_rating']=IMDb_rating
```

```python
IMDBs_Top_rated_indian_movies
```

Let's save data in csv file:

```python
IMDBs_Top_rated_indian_movies.to_csv('IMDBs_Top_rated_indian_movies.csv')
```
