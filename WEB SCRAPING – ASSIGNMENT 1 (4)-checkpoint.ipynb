4. Write a python program to scrap book name, author name, genre and book review of
any 5 books from ‘www.bookpage.com’

```python
#importing the required libraries
from bs4 import BeautifulSoup
import requests
```

First we will send get request to the webserver to get the source code of the page which is to be scraped by using the request library

```python
page=requests.get("https://bookpage.com/reviews")
page
```

```python
page.content
```

let's see the page content with the help of beautiful soup.

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify()) #see the content with alingment
```

```python
soup.title 
```

```python
book=soup.find_all('h4', class_="italic")#extracting book name
author=soup.find_all('p', class_='sans bold')#extracting book's author
review=soup.find_all('p', class_="excerpt")#extracting book's view
```

```python
#extracting only 5 books name
b=[]
for i in book:
        b.append(i.get_text().replace('\n',''))

book_name= b[0:5]
book_name
```

```python
#extracting only 5 books author name
a=[]
for i in author:
        a.append(i.get_text().replace('\n',''))

author_name= a[0:5]
author_name
```

```python
#extracting only 5 books review
r=[]
for i in review: 
    if i is None:
        r.append(i.p.get_text().replace('\n',''))
    else:
        r.append(i.get_text().replace('\n',''))
    
reviews= r[0:5]
reviews
```

Let's create dataframe and store extracted data in table format:

```python
import pandas as pd
bookpage=pd.DataFrame({})
bookpage['book_name']= book_name
bookpage['author_name']=author_name
bookpage['reviews']=reviews
```
