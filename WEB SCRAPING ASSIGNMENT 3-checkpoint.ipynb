```python
links=[]
for i in urls:
    driver.get(i)
    
    try:
        for l in driver.find_elements_by_xpath('//span[@class="rush-component"]/a'):
            links.append(l.get_attribute('href'))
    except NoSuchElementException:
        links.append("-")
```

```python
Brand_Name=[]
Name_of_the_Product=[]
Rating=[]
No_of_Ratings=[]
Price=[]
Return_Exchange=[]
Expected_Delivery=[]
Availability=[]
Other_Details=[]
Product_URL=[]

for i in links:
    driver.get(i)
    
    try:
        b = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[6]/div[4]/div[4]/div[35]/div/table/tbody/tr[1]/td[2]/span').text
        Brand_Name.append(b)
    except NoSuchElementException as e:
        Brand_Name.append(a)
        
    try:
        n = driver.find_element_by_xpath('//span[@class="a-size-large product-title-word-break"]').text
        Name_of_the_Product.append(n)
    except NoSuchElementException:
        Name_of_the_Product.append(a)
        
    try:
        r = driver.find_element_by_xpath('//*[@id="acrPopover"]/span[1]/a/i').text
        Rating.append(r)
    except NoSuchElementException:
        Rating.append(a)
        
        
    try:
        nr = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[6]/div[4]/div[4]/div[5]/div/span[3]/a/span').text
        No_of_Ratings.append(nr)
    except NoSuchElementException:
        No_of_Ratings.append(a)
        
    try:
        p = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[6]/div[4]/div[4]/div[9]/div[1]/div/table/tbody/tr[2]/td[2]').text
        Price.append(p)
    except NoSuchElementException:
        Price.append(a)
        
    try:
        re = driver.find_element_by_xpath('//*[@id="RETURNS_POLICY"]/span/div[2]/a').text
        Return_Exchange.append(re)
    except NoSuchElementException:
        Return_Exchange.append(a)
        
    try:
        ed = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[6]/div[4]/div[4]/div[11]/div').text
        Expected_Delivery.append(ed)
    except NoSuchElementException:
        Expected_Delivery.append(a)
        
    try:
        av = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[6]/div[4]/div[4]/div[21]/div[1]').text
        Availability.append(av)
    except NoSuchElementException:
        Availability.append(a.replace("\n"," "))
        
    try:
        od = driver.find_element_by_xpath('//*[@id="productDetails_techSpec_section_1"]').text
        Other_Details.append(od.replace("\n",", "))
    except NoSuchElementException:
        Other_Details.append(a)
```

```python
links=[]
try:
    for l in driver.find_elements_by_xpath(f'/html/body/div[6]/div[2]/div[2]/div/a'):
        links.append(l.get_attribute('href'))
except NoSuchElementException:
    links.append("-")
```

```python
links
```

```python
links=[]
for i in range(3, 14):
    try:
        for l in driver.find_elements_by_xpath(f'/html/body/div[6]/div[2]/div[2]/div[{i}]/a'):
            links.append(l.get_attribute('href'))
    except NoSuchElementException:
        links.append("-")
    
links
```

```python
a='-'
Brand_Name=[]
Model_name=[]
OS=[]
Display=[]
Processor=[]
Price=[]
Brief_description=[]

for i in links:
    driver.get(i)
    
    
    try:
        b = driver.find_element_by_xpath('/html/body/div[7]/div/div/div[1]/h1').text
        br=b.split(" ")
        Brand_Name.append(br[0])
    except NoSuchElementException as e:
        Brand_Name.append(a)
        
    try:
        sn = driver.find_element_by_xpath('/html/body/div[7]/div/div/div[1]/h1').text
        Model_name.append(sn)
    except NoSuchElementException:
        Model_name.append(a)
        
    try:
        c = driver.find_element_by_xpath('/html/body/div[7]/div/div/div[6]/div/div[2]/div[3]/div/ul/li[1]/div').text
        OS.append(c)
    except NoSuchElementException:
        OS.append(a)
        
        
    try:
        ram = driver.find_element_by_xpath('/html/body/div[7]/div/div/div[6]/div/div[2]/div[3]/div/ul/li[2]/div').text
        Display.append(ram)
    except NoSuchElementException:
        Display.append(a)
        
    try:
        p = driver.find_element_by_xpath('/html/body/div[7]/div/div/div[6]/div/div[2]/div[3]/div/ul/li[3]/div').text
        Processor.append(p)
    except NoSuchElementException:
        Processor.append(a)
        
    try:
        re = driver.find_element_by_xpath('//i[@class="fa fa-inr"]').text
        Price.append(re)
    except NoSuchElementException:
        Price.append(a)
        
    try:
        ed = driver.find_element_by_xpath('//*[@id="MoreLink_content-container-new"]/p').text
        Brief_description.append(ed)
    except NoSuchElementException:
        Brief_description.append(a)
```

```python
top_10_laptops=pd.DataFrame({})
top_10_laptops['Brand_Name']=Brand_Name
top_10_laptops['Model_name']=Model_name
top_10_laptops['OS']=OS
top_10_laptops['Display']=Display
top_10_laptops['Processor']=Processor
top_10_laptops['Price']=Price
top_10_laptops['Brief_description']=Brief_description

top_10_laptops
```
