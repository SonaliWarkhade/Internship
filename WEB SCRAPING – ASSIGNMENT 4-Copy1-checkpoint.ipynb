2. Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.

You need to find following details:
A) Match title (I.e. 1st ODI)
B) Series
C) Place
D) Date
E) Time

Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
#webdriver
driver=webdriver.Chrome(r"C:\Users\SAGAR KADAM\Downloads\chromedriver_win32 (1)\chromedriver.exe")
```

```python
#opening the webpage
driver.get('https://www.bcci.tv/')
```

```python
international=driver.find_element_by_xpath('/html/body/div[3]/div/div[2]/div[2]/nav/ul/li[1]/div[2]')
fixtures=driver.find_element_by_xpath('/html/body/div[3]/div/div[2]/div[2]/nav/ul/li[1]/div[2]/div/ul/li[1]/a')
```

```python
action= ActionChains(driver)
action.move_to_element(international).move_to_element(fixtures).click().perform()
```

```python
match_title=[]
series=[]
place=[]
date=[]
time=[]

m_t=driver.find_elements_by_xpath('//strong[@class="fixture__name fixture__name--with-margin"]')
for i in m_t:
    match_title.append(i.text)   
    
s=driver.find_elements_by_xpath('//span[@class="u-unskewed-text fixture__tournament-label u-truncated"]')
for i in s:
    series.append(i.text)

p=driver.find_elements_by_xpath('//p[@class="fixture__additional-info"]/span[1]')
for i in p:
    place.append(i.text)
    
day=driver.find_elements_by_xpath('//span[@class="fixture__day"]')
d=driver.find_elements_by_xpath('//span[@class="fixture__date"]')
m=driver.find_elements_by_xpath('//span[@class="fixture__month"]')
for i, j, k in zip(day, d, m):
    date.append(i.text + " " + j.text + " " + k.text)

t=driver.find_elements_by_xpath('//span[@class="fixture__time"]')
for i in t:
    time.append(i.text)
```

```python
international_fixtures=pd.DataFrame()
international_fixtures["Match Title"]=match_title
international_fixtures['Series']=series
international_fixtures['Place']=place
international_fixtures['Date']=date
international_fixtures['Time']=time
international_fixtures
```

5. Scrape the details of trending repositories on Github.com.
Url = https://github.com/
    
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used

Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
#webdriver
driver=webdriver.Chrome(r"C:\Users\SAGAR KADAM\Downloads\chromedriver_win32 (1)\chromedriver.exe")
```

```python
#opening the webpage
driver.get('https://github.com/trending')
```

```python
links=[]
try:
    for l in driver.find_elements_by_xpath('//h1[@class="h3 lh-condensed"]/a'):
        links.append(l.get_attribute('href'))
except NoSuchElementException:
        links.append("-")
links
```

```python
Repository_title=[]
Repository_description=[]
Contributors_count=[]
Language_used=[]
a="--"

for i in links:
    driver.get(i)
    
    #scraping the titles
    try:
        t = driver.find_element_by_xpath('//*[@id="js-repo-pjax-container"]/div[1]/div[1]/div/h1').text
        Repository_title.append(t.replace("\n",""))
    except NoSuchElementException:
        Repository_title.append(a)

    #scraping the description
    try:
        d = titles = driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]/div/div[1]/div/p').text
        Repository_description.append(d) 
    except NoSuchElementException:
        Repository_description.append(a)    
                        
    #scraping the company_name 
    try:
        #c = driver.find_element_by_xpath('//div[@class="flex-shrink-0 col-12 col-md-3"]/div/div[3]/div/h2/a/span').text
        c= driver.find_element_by_xpath('//div[@class="BorderGrid BorderGrid--spacious"]/div[2]/div/h2/a/span').text
        Contributors_count.append(c) 
    except NoSuchElementException:
        Contributors_count.append(a)

        
    #scraping the Language
    try:
        l1 = driver.find_element_by_xpath('//a[@class="d-inline-flex flex-items-center flex-nowrap Link--secondary no-underline text-small mr-3"]').text
        Language_used.append(l1.replace("\n"," - ")) 
    except NoSuchElementException:
        Language_used.append(a)
```

```python
print(len(Repository_title),len(Repository_description),len(Contributors_count),len(Language_used))
```

```python
Github=pd.DataFrame({"Repository title":Repository_title, "Repository description":Repository_description,
                    "Contributors count":Contributors_count,"Language used":Language_used})
Github
```
